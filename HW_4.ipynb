## Задача 3.

#### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
#### Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
import numpy as np
```

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
import seaborn as sns
```

```python
plt.figure(figsize = (10, 8))
sns.heatmap(corr_matrix, annot = True, linewidths = 0.5, cmap = 'coolwarm');
```

##### Наиболее сильное влияние на стоимость недвижимости оказывают количество жилых метров и качество жилья (коэффициент 0.7), чем выше качество жилья и чем больше количество жилых метров, тем дороже объект недвижимости. Также стоимость жилья находится в прямой зависимости от количества квадратных метров, расположенных над землей (коэффициент 0.6), и количества ванных комнат (коэффициент 0.5). Кроме того, стоимость жилья напрямую зависит от вида (коэффициент 0.4), количества спален, количества этажей, наличия вида на набережную, количества квадратных метров под землей и географического расположения (широта) (последние пять характеристик жилья имеют коэффициент 0.3).

```python
sns.jointplot(x = df['sqft_living'], y = df['price'], kind = 'reg');
```

##### Чем больше количество жилых квадратных метров имеет дом, тем выше его стоимость.

```python
sns.jointplot(x = df['grade'], y = df['price'], kind = 'reg');
```

###### Чем выше оценка качества дома, тем выше его стоимость.
